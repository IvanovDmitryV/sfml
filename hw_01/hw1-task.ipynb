```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO

def grp_lmt_b(s):
    if s<=10000            : return 'A'
    if 10000  < s <= 100000: return 'B'
    if 100000 < s <= 200000: return 'C'
    if 200000 < s <= 400000: return 'D'
    if 400000 < s <= 700000: return 'E'
    if 700000 < s          : return 'F'

df['NewColumn'] = df['LIMIT_BAL'].map(grp_lmt_b)
#df[['LIMIT_BAL','NewColumn']].head(15)
```

```python
# Потому построил гистограмму средних LIMIT_BAL от EDUCATION в разбивке по SEX
mean_educ_ml =[]
mean_educ_fml =[]
loc_ml=[]
loc_fml=[]
education = range(0,7)
wdtf = 0.3
for i in education:
    mean_educ_ml.append(df[df['SEX'].isin([1]) & df['EDUCATION'].isin([i])]['LIMIT_BAL'].mean())
    loc_ml.append(education[i])
    mean_educ_fml.append(df[df['SEX'].isin([2]) & df['EDUCATION'].isin([i])]['LIMIT_BAL'].mean())
    loc_fml.append(education[i]+wdtf)
fig,ax= plt.subplots(figsize=(20,7))
plt.bar(loc_ml, mean_educ_ml, width=wdtf)
plt.bar(loc_fml, mean_educ_fml, width=wdtf)
plt.legend(['male', 'female'])
```

```python
# ну на всякий случай построил необработанную зависимость кредитного лимита и образования в разбивке по SEX
fig,ax= plt.subplots(figsize=(20,7))
plt.plot(
        df[df['SEX'].isin([1])]['EDUCATION']-0.05, df[df['SEX'].isin([1])]['LIMIT_BAL'],'b^',
        df[df['SEX'].isin([2])]['EDUCATION']+0.05, df[df['SEX'].isin([2])]['LIMIT_BAL'],'r+'
        )
plt.legend(['male', 'female'])
```
